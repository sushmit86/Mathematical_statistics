```python
from bokeh.plotting import figure, show, output_notebook
import pandas as pd
import numpy as np

from statsmodels import stats as sm
```

```python
mu, sigma = 25, 4 # mean and standard deviation
```

```python
output_notebook()
```

```python
counter = 0
p = figure(width=400, height=400)
for i in range(1000):
    x = np.random.normal(mu,sigma,30)
    L = np.mean(x)  - 1.96 *4/(np.sqrt(30))
    U = np.mean(x)  + 1.96 *4/(np.sqrt(30))
    if (L < 25 and U > 25):
        counter = counter + 1
    if(i<100):
        p.segment(L,i,U,i,color = 'Green')
print(counter/1000)
p.segment(25,0,25,100, color = 'red')
show(p)
```
